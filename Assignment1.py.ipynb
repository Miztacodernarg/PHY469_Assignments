```python
#ASSIGNMENT

# Program to calculate the specific heat capacity of a solid

import numpy as np
import matplotlib.pyplot as plt

# Parameters for calculating heat capacity

H_cap = 100  # Heat capacity of the solid
T_ini = 20  # Initial temperature of the solid
T_E = 30  # Temperature of the environment
k_const = 0.5  # Thermal conductivity of the solid
Area = 2  # Surface area of the solid
d = 0.1  # Thickness of the solid
Vol = Area * d  # Volume of the solid

# The time step and range
dt = 0.1  # Time step
t = np.arange(0, 10, dt)  # Time range from 0 to 10 with a time step of dt

# An array for temperature
T = np.zeros_like(t)
T[0] = T_ini

# A for loop that solve the equation numerically
for i in range(1, len(t)):
    dQdt = k_const * Area * (T_E - T[i-1]) / d  # Calculating the rate of heat transfer
    dT = dQdt / (H_cap * Vol)  # Calculating the change in temperature
    T[i] = T[i-1] + dT  # Update the temperature
    print('T_change = {:.3f}K;'.format(dT))

# A plot for the temperature as a function of time
plt.plot(t, T)
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature of the Solid over Time')
plt.show()

# This part calculates the specific heat capacity
heat_input = np.arange(0, 1000, 100)  # the range of heat inputs
temperature_change = np.zeros_like(heat_input)

for i in range(len(heat_input)):
    dQ = heat_input[i]
    dT = dQ / (H_cap * Vol)
    temperature_change[i] = dT

plt.plot(heat_input, temperature_change)
plt.xlabel('Heat Input')
plt.ylabel('Temperature Change')
plt.title('Temperature Change vs. Heat Input')
plt.show()


# Calculating the slope of the plot to get specific heat capacity
slope = np.gradient(temperature_change, heat_input)
specific_heat_capacity = np.mean(slope)

print("The specific heat capacity of the solid is:", specific_heat_capacity)
```
